```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 표준 정규분포의 x 값 범위 정의
x = np.linspace(-5, 5, 1000)

# 표준 정규분포의 확률밀도함수(PDF) 값 계산
y = norm.pdf(x, 0, 1)

# 플롯 생성
plt.plot(x, y)

# 타이틀 및 레이블 추가
plt.title('Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')

# 그래프 표시
plt.show()
```
