```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# 균등 분포의 x 값 범위 정의
x = np.linspace(-2, 2, 1000)

# 균등 분포의 확률밀도함수(PDF) 값 계산
y = uniform.pdf(x, loc=-1, scale=2)  # loc은 시작점, scale은 범위의 길이

# 플롯 생성
plt.plot(x, y)

# 타이틀 및 레이블 추가
plt.title('Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')

# 그래프 표시
plt.show()
```
